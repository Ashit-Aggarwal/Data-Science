**Obeservation:**
There is a big difference between 75% and max values for residual sugar, chlorides and free sulphur dioxide

```python
fig, ax= plt.subplots(ncols=3,figsize=(14,7))
sns.boxplot(ax=ax[0], y=df['residual sugar'])
sns.boxplot(ax=ax[1], y=df['chlorides'])
sns.boxplot(ax=ax[2], y=df['free sulfur dioxide'])
# ax[0].set_xlabel()
plt.show()
```

```python
df2=df[~((df['residual sugar']>4) | (df['chlorides']>0.3) | (df['free sulfur dioxide']>60))]
```

```python
df2.describe()
```

```python
plt.figure(figsize=(24,24))

cols=df2.columns.values
for i in range(0,len(df.columns)):
#     plt.subplot(3,4,i+1)
    sns.displot(x=df2[cols[i]], kde=True) 
# plt.show()
```

only residual sugar and pH seems to be right probability distrivution curves

- **From above we see that when fixed acidity is between around 8-10 and density is between 0.993 and 0.996 we had max quality.**
- **Also when fixed acidity is between around 7-10 and citric acid is between 0.3 and 0.5 we had max quality**
- **Alcohol is bewtween 11.5 nd 12.5 and density is between 0.993 and 0.998**
- **Citric acid is between 0.3 and 0.6 and pH is between 3.2 and 3.4**

```python
data=df2[(df2['fixed acidity']> 7.5) & (df2['fixed acidity']<10) & (df2['density']>0.993) & (df2['density']<0.998) & (df2['citric acid']>0.3) & (df2['citric acid']<0.6)
        & (data['alcohol']>11.5) & (data['alcohol']<12.5)& (data['pH']>3.2) & (data['pH']<3.4)]
```
