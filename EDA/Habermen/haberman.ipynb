## Haberman Data Info

**Dataset Link:** https://www.kaggle.com/gilsousa/habermans-survival-data-set
        
**Relevant Information:**
The dataset contains cases from a study that was conducted between
1958 and 1970 at the University of Chicago's Billings Hospital on
the survival of patients who had undergone surgery for breast
cancer.

Number of Instances: 306

Number of Attributes: 4 (including the class attribute)

**Attribute Information:**

Age of patient at time of operation (numerical)
Patient's year of operation (year - 1900, numerical)
Number of positive axillary nodes detected (numerical)
Survival status (class attribute)
1 = the patient survived 5 years or longer
2 = the patient died within 5 year

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('haberman.csv', header=None)  # Header is not there 
```

#### Setting up Column Names

```python
df.columns=['Age','opYear','auxNode','Survival']
```

#### Getting basic info

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

- There are no null vales in the 306 records
- There are atleast 25% patients with 0 Auxillary nodes 
- Median age = 52
- Median for Auxillary node = 1. There were a fe patients with very high positive auxillary node

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(y=df.auxNode)
```

```python
sns.displot(data=df, x='Survival')
```

```python
df.Survival.value_counts(normalize=True)
```

**We can see that nearly 74% of all patients survived more than 5 years**

```python
df.auxNode.value_counts(normalize=True).head(10)
```

```python
sns.countplot(x=df.auxNode)
```

```python
sns.set_style('whitegrid')
sns.scatterplot(data=df,x='Age',y='auxNode')
plt.show()
```

- Maximum positive auxillary nodes observed between 50-55 age

```python
sns.boxplot(data=df,y='Age',x='Survival')
plt.show()
```

- Survival less than 5 years has slightly more median than survival less than 5 years.
- Generally age of patients is between 45-60

#### Does year of operation has any relation with survival?

```python
sns.boxplot(data=df[df['auxNode']!=0], x='Survival', y='opYear')
plt.show()
```

```python
sns.boxplot(data=df[df['auxNode']!=0], hue='Survival', x='opYear',y='Age')
plt.show()
```

**There is no relation between year of operation and survival rate. IQR and Median is nearly the same**

```python
sns.scatterplot(data=df, y='auxNode',x='Age',hue='Survival', palette=['orange','blue'])
plt.show()
```

**More deaths within 5 year of operation above age of 45 observed**

```python
sns.boxplot(data=df, x='Survival', y='auxNode')
```

**More the positive Auxillary node more the chances of deaths within 5 years of operation**

```python
sns.heatmap(df.corr(), annot=True, cmap='RdYlGn')
```

**No strong relation observed within parameters**

```python
sns.pairplot(df)
```
