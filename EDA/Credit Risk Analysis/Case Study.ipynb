#### We will now impute all the variables which starts with AMT_REQ_CREDIT_BUREAU

```python
# Credit bureau parameters are integers. check the frequency of the integers for all the relevant columns for amt_req_credit_bureau

new_app[['AMT_REQ_CREDIT_BUREAU_HOUR','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK',
          'AMT_REQ_CREDIT_BUREAU_MON','AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR']].apply(pd.value_counts)
```

#### We will use median method for imputation as missing values are relatively lower and for all the categories 0 is the mode as well

```python
# We can impute these columns with median as 0 has the highest count and 50th percentile for these columns is 0 as well

amedian = new_app[['AMT_REQ_CREDIT_BUREAU_HOUR','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK',
          'AMT_REQ_CREDIT_BUREAU_MON','AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR']].median()[0]
new_app.AMT_REQ_CREDIT_BUREAU_HOUR.fillna(amedian, inplace = True)
new_app.AMT_REQ_CREDIT_BUREAU_DAY.fillna(amedian, inplace = True)
new_app.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(amedian, inplace = True)
new_app.AMT_REQ_CREDIT_BUREAU_MON.fillna(amedian, inplace = True)
new_app.AMT_REQ_CREDIT_BUREAU_QRT.fillna(amedian, inplace = True)
new_app.AMT_REQ_CREDIT_BUREAU_YEAR.fillna(amedian, inplace = True)
new_app.iloc[:,-6:].isnull().sum() #check the null values to ensure imputation is correct and complete
```

```python
#check the variable type for the remaining columns

new_app[new_app.columns[(new_app.isnull().sum()) > 0]].head()
```

```python
# Since we are going to run multiple barcharts for multivariate analysis, we will define a function which will help us in...
# ... running the charts with 1 line codes

def Barplot(xvar,yvar,huecol=None):

    if huecol==None:
        group_var_list=[xvar]
    else:
        group_var_list=[xvar,huecol]

    res1=new_app.groupby(by=group_var_list)[yvar].value_counts(normalize=True).mul(100).rename('Percent').reset_index()
    res2=new_app.groupby(by=group_var_list)[yvar].value_counts().rename('Count').reset_index()  
#         res3=new_app.groupby(by=group_var_list)[yvar].value_counts(normalize=True).mul(100).rename(rename_var).reset_index()
#         print(res3.head(20))

    #Plotting the Data
    sns.set(style="whitegrid")
    fig,ax =plt.subplots(2,2, figsize=(15,15))
    
    ## Create Plot for DataFrame where Target=0 i.e. Defaulter is not present
    sns.barplot(data=res1[res1['TARGET']==0], x=xvar,y='Percent',hue=huecol,ax=ax[0,0])
    ax[0,0].set_title('Non-Defaulter',fontdict={'fontsize': 20, 'fontweight' : 5})
    ax[0,0].set_ylabel(yvar + ' Percent', fontdict={'fontsize': 15, 'fontweight' : 5})
    ax[0,0].set_xlabel('')
    ax[0,0].set_xticklabels('')
#     ax0.set_xlabel(xvar,fontdict={'fontsize':15, 'fontweight': 5})
#     ax0.set_xticklabels(ax0.get_xticklabels(),rotation=90)
    ax[0,0].legend().set_visible(False)


    ## Create Plot for DataFrame where Target=1 i.e. Defaulter is present
    
    sns.barplot(data=res1[res1['TARGET']==1], x=xvar,y='Percent',hue=huecol,ax=ax[0,1])
    ax[0,1].set_title('Defaulter',fontdict={'fontsize': 20, 'fontweight' : 5})
    ax[0,1].set_xlabel('')
    ax[0,1].set_ylabel('')
    ax[0,1].set_xticklabels('')
#     ax1.set_xlabel(xvar,fontdict={'fontsize':15, 'fontweight': 5})
#     ax1.set_xticklabels(ax1.get_xticklabels(),rotation=90)
    ax[0,1].legend(bbox_to_anchor=(1,1),loc="upper left")
    
    sns.barplot(data=res2[res2['TARGET']==0], x=xvar,y='Count',hue=huecol,ax=ax[1,0])
    ax[1,0].set_ylabel(yvar + ' Count', fontdict={'fontsize': 15, 'fontweight' : 5})
    ax[1,0].set_xlabel(xvar,fontdict={'fontsize':15, 'fontweight': 5})
    ax[1,0].set_xticklabels(ax[1,0].get_xticklabels(),rotation=90)
    ax[1,0].legend().set_visible(False)


    ## Create Plot for DataFrame where Target=1 i.e. Defaulter is present
    
    sns.barplot(data=res2[res2['TARGET']==1], x=xvar,y='Count',hue=huecol,ax=ax[1,1])
    ax[1,1].set_title('Defaulter',fontdict={'fontsize': 20, 'fontweight' : 5})
    ax[1,1].set_ylabel('')
    ax[1,1].set_xlabel(xvar,fontdict={'fontsize':15, 'fontweight': 5})
    ax[1,1].set_xticklabels(ax[1,1].get_xticklabels(),rotation=90)
    ax[1,1].legend(bbox_to_anchor=(1,1),loc="upper left")
    

    plt.show()
    
```

## Merging Application and Previous application Datasets

Perform Analysis only on records present in new and previous application Datasets.

```python
# Doing inner join here as we need to find relation between common records for new and previous applications only.
# Left and right joins are not used as both the datasets have been already analyzed separately

merged_df=pd.merge(new_app,
         prev_app[['SK_ID_CURR','NAME_CONTRACT_STATUS','AMT_ANNUITY','AMT_APPLICATION','AMT_CREDIT','AMT_DOWN_PAYMENT','AMT_GOODS_PRICE'
                  ,'NAME_CONTRACT_TYPE','NAME_CLIENT_TYPE','NAME_GOODS_CATEGORY','NAME_PORTFOLIO','NAME_PRODUCT_TYPE',
                   'CHANNEL_TYPE','SELLERPLACE_AREA','NAME_SELLER_INDUSTRY','NAME_CASH_LOAN_PURPOSE','CODE_REJECT_REASON',
                  'NAME_YIELD_GROUP','PRODUCT_COMBINATION']],
         how='inner',
         on='SK_ID_CURR'
        
        )

merged_df.head()
```

```python
def Barplot2(xvar,yvar,huecol=None):

    if huecol==None:
        group_var_list=[xvar]
    else:
        group_var_list=[xvar,huecol]

    res1=merged_df.groupby(by=group_var_list)[yvar].value_counts(normalize=True).mul(100).rename('Percent').reset_index()
    res2=merged_df.groupby(by=group_var_list)[yvar].value_counts().rename('Count').reset_index()  
#         res3=new_app.groupby(by=group_var_list)[yvar].value_counts(normalize=True).mul(100).rename(rename_var).reset_index()
#         print(res3.head(20))

    #Plotting the Data
    sns.set(style="whitegrid")
    fig,ax =plt.subplots(2,2, figsize=(15,15))
    
    ## Create Plot for DataFrame where Target=0 i.e. Defaulter is not present
    sns.barplot(data=res1[res1['TARGET']==0], x=xvar,y='Percent',hue=huecol,ax=ax[0,0])
    ax[0,0].set_title('Non-Defaulter',fontdict={'fontsize': 20, 'fontweight' : 5})
    ax[0,0].set_ylabel(yvar + ' Percent', fontdict={'fontsize': 15, 'fontweight' : 5})
    ax[0,0].set_xlabel('')
    ax[0,0].set_xticklabels('')
#     ax0.set_xlabel(xvar,fontdict={'fontsize':15, 'fontweight': 5})
#     ax0.set_xticklabels(ax0.get_xticklabels(),rotation=90)
    ax[0,0].legend().set_visible(False)


    ## Create Plot for DataFrame where Target=1 i.e. Defaulter is present
    
    sns.barplot(data=res1[res1['TARGET']==1], x=xvar,y='Percent',hue=huecol,ax=ax[0,1])
    ax[0,1].set_title('Defaulter',fontdict={'fontsize': 20, 'fontweight' : 5})
    ax[0,1].set_xlabel('')
    ax[0,1].set_ylabel('')
    ax[0,1].set_xticklabels('')
#     ax1.set_xlabel(xvar,fontdict={'fontsize':15, 'fontweight': 5})
#     ax1.set_xticklabels(ax1.get_xticklabels(),rotation=90)
    ax[0,1].legend(bbox_to_anchor=(1,1),loc="upper left")
    
    sns.barplot(data=res2[res2['TARGET']==0], x=xvar,y='Count',hue=huecol,ax=ax[1,0])
    ax[1,0].set_ylabel(yvar + ' Count', fontdict={'fontsize': 15, 'fontweight' : 5})
    ax[1,0].set_xlabel(xvar,fontdict={'fontsize':15, 'fontweight': 5})
    ax[1,0].set_xticklabels(ax[1,0].get_xticklabels(),rotation=90)
    ax[1,0].legend().set_visible(False)


    ## Create Plot for DataFrame where Target=1 i.e. Defaulter is present
    
    sns.barplot(data=res2[res2['TARGET']==1], x=xvar,y='Count',hue=huecol,ax=ax[1,1])
    ax[1,1].set_title('Defaulter',fontdict={'fontsize': 20, 'fontweight' : 5})
    ax[1,1].set_ylabel('')
    ax[1,1].set_xlabel(xvar,fontdict={'fontsize':15, 'fontweight': 5})
    ax[1,1].set_xticklabels(ax[1,1].get_xticklabels(),rotation=90)
    ax[1,1].legend(bbox_to_anchor=(1,1),loc="upper left")
    

    plt.show()
    
```
