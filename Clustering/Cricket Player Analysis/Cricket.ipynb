```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.neighbors import NearestNeighbors
```

```python
df=pd.read_csv('Cricket.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
scaler=StandardScaler()
```

```python
df.dtypes!=object
```

```python
col=df.dtypes[df.dtypes!=object].reset_index()
col=col.iloc[:,0]
col
```

```python
cric=pd.DataFrame()
```

```python
col=['SR','Ave']
```

```python
cric=scaler.fit_transform(df[col])
```

```python
cric
```

```python
kmeans=KMeans(n_clusters=4,random_state=100,max_iter=100)
ml=kmeans.fit(cric)
```

```python
ml.labels_
```

```python
df['cluster']=ml.labels_
```

```python
df[df.Player.str.startswith('V ')]
```

```python
df[df['cluster']==3]
```

```python
df[df['cluster']==0]
```

```python
df[df['cluster']==2]
```

```python
sns.scatterplot(df.SR,df.Ave)
plt.show()
```

## Hirerachical Clustering

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
mergings=linkage(cric,method='complete',metric='euclidean')
```

```python
dendrogram(mergings)
plt.show()
```

```python
label=cut_tree(mergings,n_clusters=4).reshape(-1,)
```

```python
df['labels']=label
```

```python
df[df['Player'].str.startswith('V ')]
```

```python
df[df['labels']==3]
```
